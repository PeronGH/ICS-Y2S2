```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G1 = nx.DiGraph()
G1.add_edges_from([['a', 'd'], ['a', 'e'], ['d', 'b'],
                   ['b', 'c'], ['e', 'f'], ['f', 'c']])

nx.draw(G1, with_labels=True)
plt.show()
```

```python
adjacency_matrix = nx.adjacency_matrix(G1)
print(adjacency_matrix.todense())

incidence_matrix = nx.incidence_matrix(G1).astype(int)
print(incidence_matrix.todense())
```

```python
# Define the binary tree
tree = nx.balanced_tree(2, 3)

# Define the positions of the nodes
pos = {0: (0.5, 1.0)}
y_pos = 0.8
for level in range(1, 4):
    num_nodes = 2 ** level
    x_pos = 0.5 - (num_nodes / 2) * 0.1
    for i in range(num_nodes):
        pos[num_nodes + i - 1] = (x_pos + i * 0.1, y_pos)
    y_pos -= 0.2

# Draw the tree
nx.draw(tree, pos=pos, with_labels=True, node_size=500, font_size=16)

# Show the image
plt.show()
```
